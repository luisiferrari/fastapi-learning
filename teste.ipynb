```python
def bubble_sort(lista):
    n = len(lista)
    # O loop externo garante que passaremos pela lista n vezes
    for i in range(n):
        # O loop interno compara os elementos vizinhos
        # n - i - 1 evita que olhemos para o que já está ordenado no final
        for j in range(0, n - i - 1):
            if lista[j] > lista[j + 1]:
                # Troca os elementos de lugar (Swap)
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
    return lista

# Testando com a sua lista:
minha_lista = [7, 2, 9, 0, 5, 1, 8, 4, 6, 3]
lista_ordenada = bubble_sort(minha_lista)
print(lista_ordenada)
```
